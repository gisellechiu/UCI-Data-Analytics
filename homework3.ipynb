
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

all = df.groupby('Gender')

count = all['SN'].count()
countdf = pd.DataFrame(count)

avg = all['Price'].mean()

totalpurchase = all['Price'].sum()


norm_total = totalpurchase / count


countdf['Average Purchase Price'] = avg
purchasedf = pd.DataFrame(countdf)

newdf = purchasedf.rename(columns={'SN': 'Purchase Count'})


newdf['Total Purchase Price'] = totalpurchase
purchasedf = pd.DataFrame(newdf)


purchasedf['Normalized Totals'] = norm_total
purchasedf = pd.DataFrame(purchasedf)
purchasedf
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

df['Total Count'] = pd.cut(df["Age"], age_bins, labels=group_names)
groups = df.groupby('Total Count')
groupsdf = groups.count()


agedf = groupsdf[["Age"]]
agedf
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

SN = df.groupby('SN')['Price'].count()
SNdf = pd.DataFrame(SN)


uniquedf = df.groupby('SN')['Price'].sum()


avgSN = df.groupby('SN')['Price'].mean()


SNdf['Average Purchase Price'] = avgSN


SNdf['Total Purchase Value'] = uniquedf
SNdf


updated = SNdf.rename(columns={'Price': 'Purchase Count'})


sorted = updated.sort_values(by=['Total Purchase Value'], ascending=False)
sorted


sorted.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

item = df.groupby('Item ID')['Price'].count()
itemdf = pd.DataFrame(item)


unique = df.groupby('Item ID')['Price'].sum()


name = df.iloc[:, 3]
itemdf['Item Name'] = name


item_price_df = df.iloc[:, 4]
itemdf['Item Price'] = item_price_df


itemdf['Total Purchase Value'] = unique
itemdf


updated = itemdf.rename(columns={
    'Price': 'Purchase Count',
})


organized = updated[['Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]


sorted_items_df = organized.sort_values(by=['Purchase Count'], ascending=False)


sorted_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python

item2 = df.groupby('Item ID')['Price'].count()
item2df = pd.DataFrame(item2)

# Calculate 'Total Purchase Price' of each item
unique2 = df.groupby('Item ID')['Price'].sum()

# Add 'Item Name' to table
name2 = df.iloc[:, 3]
item2df['Item Name'] = name2

# Add 'Item Price' to table
price2 = df.iloc[:, 4]
item2df['Item Price'] = price2

# Add 'Total Purchase Value' to table
item2df['Total Purchase Value'] = unique2
item2df

# Rename 'Price' to 'Purchase Count'
updated2 = item2df.rename(columns={
    'Price': 'Purchase Count',
})

# Rearrange columns
organized2 = updated2[['Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]

# Sort 'Purchase Count' from highest to lowest
sorted_items2_df = organized2.sort_values(by=['Total Purchase Value'], ascending=False)

# Show top 5 popular items only
sorted_items2_df.head()
```
